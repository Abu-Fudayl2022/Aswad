## Creating a line plot

This tutorial shows how to create a line plot

```python
x = [1, 2, 3, 4]
y = [-1, -2, -3, -4]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```
